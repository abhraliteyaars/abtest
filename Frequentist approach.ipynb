For the next set of analysis we will divide the dataframe into Control and Experience Dataframes

```python
df_ctrl=df_new[df_new['Search activity- Control']==1]
df_exp=df_new[df_new['Search activity- Experience']==1]
```

```python
# no of visitors in each experience
ctrl_visitors=df_ctrl['CustID'].nunique()
exp_visitors=df_exp['CustID'].nunique()
print('ctrl_visitors',ctrl_visitors,'     exp_visitors',exp_visitors)
```

```python
#to ensure there is balance in the populations I am sampling 500,000 line items from both sets
df_ctrl=df_ctrl.sample(500000)
df_exp=df_exp.sample(500000)
exp_n=ctrl_n=500000
```

```python
# no of visitors in each experience

ctrl_visitors=df_ctrl['CustID'].nunique()
exp_visitors=df_exp['CustID'].nunique()
print('ctrl_visitors',ctrl_visitors,'     exp_visitors',exp_visitors)
```

```python
# no of searches in each experience
ctrl_search=df_ctrl['Search'].sum()
exp_search=df_exp['Search'].sum()
                  
print('ctrl_search',ctrl_search,'     exp_search',exp_search)
```

```python
#Search rate in each experience

ctrl_search_rate=ctrl_search/ctrl_visitors
exp_search_rate=exp_search/exp_visitors
lift_search=(exp_search_rate-ctrl_search_rate)/ctrl_search_rate


print('ctrl_search_rate',ctrl_search_rate,'     exp_search_rate',exp_search_rate,'     lift_search',lift_search)
```

```python
#Confidence of search rate- will do a hypothesis test

#create the contingency table
search=[exp_search_rate,ctrl_search_rate]
samples=[exp_n,ctrl_n]

z, pval = proportions_ztest(search, samples,alternative='two-sided')
z,pval
```

```python
# no of add to carts in each experience
ctrl_cart=df_ctrl['Cart Additions'].sum()
exp_cart=df_exp['Cart Additions'].sum()
                  
print('ctrl_cart',ctrl_cart,'     exp_cart',exp_cart)
```

```python
#Cart addition rate for each group
ctrl_cart_rate=ctrl_cart/ctrl_visitors
exp_cart_rate=exp_cart/exp_visitors
lift_cart=(exp_cart_rate-ctrl_cart_rate)/ctrl_cart_rate


print('ctrl_cart_rate',ctrl_cart_rate,'     exp_cart_rate',exp_cart_rate,'     lift_cart',lift_cart)
```

```python
#Confidence of add to cart rate- will do a hypothesis test

#create the contingency table
cart=[exp_cart_rate,ctrl_cart_rate]
samples=[exp_n,ctrl_n]

z, pval = proportions_ztest(cart, samples,alternative='two-sided')
z,pval
```

```python
#no of Orders in each experience
ctrl_ord=df_ctrl['Orders'].sum()
exp_ord=df_exp['Orders'].sum()
                  
print('ctrl_ord',ctrl_ord,'     exp_ord',exp_ord)
```

```python
#no of Orders in each experience
ctrl_ord_rate=ctrl_ord/ctrl_visitors
exp_ord_rate=exp_ord/exp_visitors
lift_ord=(exp_ord_rate-ctrl_ord_rate)/ctrl_ord_rate

print('ctrl_ord_rate',ctrl_ord_rate,'     exp_ord_rate',exp_ord_rate,'     lift_ord',lift_ord)
```

```python
#Confidence of order rate- will do a hypothesis test

#create the contingency table
orders=[exp_ord_rate,ctrl_ord_rate]
samples=[exp_n,ctrl_n]

z, pval = proportions_ztest(orders, samples,alternative='two-sided')
z,pval
```

As we can see none of the KPIs we were trying to see an improvement in shows any significant improvement. Recommendation is to stick to the original design
